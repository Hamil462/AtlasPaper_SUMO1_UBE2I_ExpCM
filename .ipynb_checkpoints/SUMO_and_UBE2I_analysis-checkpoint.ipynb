# Phylogenetic analysis informed by SUMO1 and UBE2I deep mutational scanning data
This notebook describes an analysis by [Jesse Bloom](http://research.fhcrc.org/bloom/en.html) of deep mutational scanning of SUMO1 and UBE2I data provided by Jochen Weile and Joe Wu in the lab of [Fritz Roth](http://llama.mshri.on.ca/).

## Main findings
Here is a summary of the main findings:

  1. The deep mutational scanning data can usefully inform phylogenetic models of orthologs of SUMO1 and UBE2I up to at least 75% protein divergence. This indicates that the experiments successfully describe at least some of the constraints on the evolution of these genes in nature.
  
  2. The best way to treat mutations that enhance activity is as deleterious.
  
  3. There are a few sites that are clearly under substantially different selection in nature versus the experiments.

## Introduction
Our overall goal is to compare the mutational effects measured in the deep mutational scanning to the constraints on the evolution of SUMO1 and UBE2I in nature. 

The approach is to use the deep mutational scanning data to create a site-specific model for how each site should evolve if the constraints in the experiments match those in nature. We then use likelihood-based phylogenetics to compare the models informed by the experiments to each and to standard models that do not use the experiments (see [Bloom 2014](http://mbe.oxfordjournals.org/content/31/8/1956) and [Bloom 2016](http://biorxiv.org/content/early/2016/01/22/037689)). These analyses are performed using the [phydms](http://jbloomlab.github.io/phydms/) software package, and allow us to assess how well the experiments capture constraints on evolution in nature.

Another question deals with how to handle mutations that enhance protein activity. There are some mutations that actually make the proteins *more* active in the lab. What are the fitness consequences of such mutations -- are then beneficial or deleterious? We can address this question by comparing models that consider activity-enhancing mutations as beneficial or deleterious.

Finally, this analysis relates to the overall question of how well experiments in the lab can be extrapolated to a wide set of orthologs. One argument is that mutational effects will shift rapidly as a protein evolves, so that measurements on one ortholog don't tell you much about other orthologs. Another argument is that mutational effects are fairly conserved, so measurements on one ortholog can be extended to others. The limited evidence so far perhaps supports this latter argument more than the former, see [Doud et al 2016](https://dx.doi.org/10.1093/molbev/msv167). However, assessing if it is useful to apply measurements on a single human ortholog of SUMO1 or UBE2I to describe orthologs from a wide range of other species will provide more insight into this issue.

## Analysis and Results

### Import modules, define some variable and functions

```python
%matplotlib inline

# import modules
import subprocess
import os
import pandas
import matplotlib
matplotlib.style.use('ggplot')
import matplotlib.pyplot as plt
import docutils.core
from IPython.display import Image, display, HTML
import Bio
import Bio.SeqIO
print("Using biopython version {0}".format(Bio.__version__))
import dms_tools
import dms_tools.file_io
print("Using dms_tools version {0}".format(dms_tools.__version__))

# genes and score files from the Roth lab
genes = ['SUMO1', 'UBE2I']

excludesites = { # exclude DMS data for positions with these Python indices from analysis
        'SUMO1':[],   # don't exclude anything
        'UBE2I':[-1], # exclude last position as DMS protein is one residue longer than ENSEMBL orthologs
        }

# perform analyses for all sequences with protein identities >= each of these cutoffs to the human DMS protein
identity_cutoffs = [0.85, 0.75]
        
def ShowPDF(pdf, width=None):
    '''Creates PNG image of *pdf* and displays inline.'''
    png = os.path.splitext(pdf)[0] + '.png'
    subprocess.check_call(['convert', '-density', '192', '-trim', pdf, png])
    display(Image(png, width=width))
```

### Process the scores to amino-acid preferences
The Roth lab provides the data from their deep mutational scanning experiments as scores. In all cases, we are using the *joint score*, which combines the result from the screen with calculations that also utilize some additional information.

We define the preference $\pi_{r,a}$ for amino-acid $a$ at site $r$ in terms of the score $s_{r,a}$ as $\pi_{r,a} = \frac{s_{r,a}}{\sum_{a'} s_{r,a'}}$. Essentially, the preferences are the scores normalized to sum to one at each site. Note that before performing this calculation we first set each score to be at least as large as a small number (0.01) to deal with negative and zero scores.

We have two sets of scores from the Roth lab: 

  1. Those calculated with conservation data
  
  2. Those calculated using no conservation data.
  
The reason for treating these separately is that there is some concern that calculations using conservation data may introduce some circularity if that data involves the sequences themselves being analyze. The scores calculated *without* conservation data therefore seem safer to use, but we will test both to see if it makes much difference.

We further transform the scores in three different ways to represent different possible biological interpretations of the fitness effects of mutations that increase activity:

  1. We just use the untransformed scores as is. This means that if a mutation increases activity in the lab, it is expected to be beneficial in nature.
  
  2. We transform any score $s_{r,a} > 1$ to a value of $s_{r,a} = 1$. This means that activity-increasing mutations are considered just as good as wildtype.
  
  3. We penalize scores > 1. Specifically, if $s_{r,a} > 1$, then we transform it to a value of $1 / s_{r,a}$. This means that mutations that increase activity relative to wildtype are construed as being *deleterious* based on the idea that both increased and decreased activity can be harmful.
  
The data output by the cell below explains how many scores are found for each data set, how many are purely interpolated versus measured, and the distribution of scores. As is obvious from the plot, the distribution of scores is bi-modal, with peaks for nearly neutral and highly deleterious mutaitons.

```python
minscore = 1e-2 # any score less than this is set to this value to avoid zero / negative values.

prefs = {} # dictionary holding names of file with preferences, keyed by *gene* and then by *scoretype* 
prefsdir = './prefs/'
if not os.path.isdir(prefsdir):
    os.mkdir(prefsdir)

scoretransforms = {
        'untransformed': lambda s: s,
        'max-scores-at-1': lambda s: min(1.0, s),
        'penalize-scores-gt1': lambda s: s if s <= 1 else 1.0 / s,
        }    
    
# get preferences from scores
for gene in genes:
    prefs[gene] = {}
    for (scoreset, filesuffix) in [('with-cons', ''), ('no-cons', '_nocons')]:
        scorefile = './scores/imputed_joint_scores_{0}{1}.csv'.format(gene, filesuffix)
        print("\nReading scores for {0} ({1} score set) from {2}".format(gene, scoreset, scorefile))
        df = pandas.read_csv(scorefile)
        print("There are {0} measured scores and {1} joint (measured+predicted) scores".format(
                df['screen.score'].count(), df['joint.score'].count()))
        df['joint.score'].plot(kind='kde', lw=2, label='{0} {1}'.format(gene, scoreset))
        sites = []
        wts = {}
        pis = dict([(scoretransform, {}) for scoretransform in scoretransforms.keys()])
        for (irow, row) in df.iterrows():
            (wt, r, mut) = (row['mut'][0], row['mut'][1 : -1], row['mut'][-1])
            if r not in wts:
                sites.append(r)
                wts[r] = wt
                for scoretransform in scoretransforms.keys():
                    pis[scoretransform][r] = {}
            for (scoretransform, lambdaf) in scoretransforms.items():
                pis[scoretransform][r][mut] = lambdaf(max(minscore, row['joint.score']))
        sites = [r for r in sites if int(r) - 1 not in excludesites[gene] + [i + len(sites) for i in excludesites[gene]]]
        for r in sites:
            for scoretransform in scoretransforms.keys():
                scoresum = float(sum(pis[scoretransform][r].values()))
                pis[scoretransform][r] = dict([(aa, pi / scoresum) for (aa, pi) in pis[scoretransform][r].items()])
        for scoretransform in scoretransforms.keys():
            fname = '{0}/{1}_{2}_{3}_prefs.txt'.format(prefsdir, gene, scoreset, scoretransform)
            print("Writing {0} preferences to {1}".format(scoretransform, fname))
            dms_tools.file_io.WritePreferences(fname, sites, wts, pis[scoretransform], None)
            prefs[gene]['{0}_{1}'.format(scoreset, scoretransform)] = fname

# distribution of scores
print("\nHere is the distribution of scores.")    
plt.legend()
plt.xlim(-1, 3)
plt.xlabel('score')
plt.show()
```

### Create alignments of SUMO1 and UBE2I orthologs
In order to compare the deep mutational scanning to selection in nature, we need a set of orthologs of each gene. I used [Ensemble protein ortholog sets](http://uswest.ensembl.org/info/genome/compara/homology_method.html), downloaded on July-14-2016. Specifically:

  1. For human SUMO1, I downloaded the coding sequences for all orthologs from [here](http://uswest.ensembl.org/Homo_sapiens/Gene/Compara_Ortholog?db=core;g=ENSG00000116030;r=2:202206180-202238608). 

  2. For human UBE2I, I downloaded the coding sequences for all orthologs from [here](http://uswest.ensembl.org/Homo_sapiens/Gene/Compara_Ortholog?db=core;g=ENSG00000103275;r=16:1308880-1327018).
  
I then further processed these orthologs with a couple of goals in mind.
  1. We can only meaningfully consider sites that are ungapped in the human orthologs, since those are the only ones for which we have deep mutational scanning data. 
  
  2. Nearly redundant sequences provide little phylogenetic information, so to speed the analysis I removed highly similar sequences. 
  
  3. It is unclear how diverged the ortholog sets should be. On the one hand, including more diverged orthologs spans a great range of evolution, so there is more data to analyze. On the other hand, if mutational effects shift among different orthologs, then including more diverged sequences will lead to analysis of proteins for which the experiments would not provide a useful description of evolutionary constraints. Therefore, for each gene I made two sets of orthologs: ones with at least 85% protein identity to the human ortholog, and ones with at least 75% identity to the human ortholog. 

The construction of these ortholog sets is done below using ``phydms_prepalignment``. The cell outputs information about the number of orthologs in each set, and plots the distribution of identities relative to the reference sequence (the human ortholog) of all sequences and the retained sequences.

```python
# get alignment of orthologs with gaps stripped relative to human protein used in DMS, filtered by protein identity
alignments = dict([(gene, {}) for gene in genes]) # keyed by gene, identity cutoff, value is file
purge = {'SUMO1':'', 'UBE2I':'Ogar'} # including Ogar somehow causes crash during tree building
for gene in genes:
    inseqs = './alignments/{0}_orthologs.fasta'.format(gene)
    for cutoff in identity_cutoffs:
        alignmentfile = './alignments/{0}_alignment_protidentity_ge_{1}.fasta'.format(gene, cutoff)
        alignments[gene][cutoff] = alignmentfile
        print("\nGetting {0} homologs with >= {1} protein identity to {2}".format(gene, cutoff, alignmentfile))
        subprocess.check_call([
                'phydms_prepalignment',
                inseqs,
                alignmentfile,
                'Hsap', # in the ortholog files, *Hsap* indicates the Homo sapiens sequence, which is our reference
                '--prealigned',
                '--minidentity', str(cutoff),
                '--minuniqueness', '2',
                '--purgeseqs', purge[gene],
                ])
        nseqs = len(list(Bio.SeqIO.parse(alignmentfile, 'fasta')))
        print("Found {0} such sequences.\nHere is a plot showing what was retained:".format(nseqs))
        plotfile = os.path.splitext(alignmentfile)[0] + '.pdf'
        ShowPDF(plotfile, width=400)
        
```

### Assess how well the evolution of each set of orthologs is described by each set of deep mutational scanning scores
We now fit phylogenetic substitution models informed by the deep mutational scanning to each set of orthologs. Recall that we have four sets of orthologs: two genes (SUMO1 and UBE2I), for each considering orthologs with at least 85% and at least 75% identities. We also have six sets of amino-acid preferences from deep mutational scanning: those calculated with and without conservation data, and those that treat activity-enhancing mutations as beneficial, equivalent to wildtype, or detrimental. In addition, for each set of preferences, we also perform a control analysis where the deep mutational scanning data (the site-specific amino-acid preferences) are *randomized* among sites. The idea here is that if the experiments describe natural evolution well because capture constraints on individual sites, then randomizing this data should abrogate any improvements due to the information captured from the experiments. Finally, we use two standard phylogenetic substitutions models: the M0 and M3 variants of the Goldman-Yang 1994 models, here denoted YNGKP to stand for the initials of the authors on the paper describing these variants [Yang et al 2000](http://www.genetics.org/content/155/1/431.short).

The actual fitting is done with the [phydms_comprehensive](http://jbloomlab.github.io/phydms/phydms_comprehensive_prog.html) program.

Note that these jobs are time consuming, so rather than running them individually we use `slurm` to submit them on the cluster. The cell below runs these jobs and prints the exact command used.

```python
print("Using " + subprocess.check_output(['phydms', '-v'], stderr=subprocess.STDOUT))

phydmsdir = './phydms_results/'
if not os.path.isdir(phydmsdir):
    os.mkdir(phydmsdir)

modelcomparisons = dict([(gene, {}) for gene in genes])
prefixes = {}
for gene in genes:
    prefixes[gene] = {}
    for cutoff in identity_cutoffs:
        alignment = alignments[gene][cutoff]
        outprefix = '{0}/{1}_identity-ge-{2}'.format(phydmsdir, gene, cutoff)
        prefixes[gene][cutoff] = outprefix
        modelcomparisons[gene][cutoff] = outprefix + '_modelcomparison.txt'
        if os.path.isfile(modelcomparisons[gene][cutoff]):
            print("\nThe output model comparison file {0} already exists. Delete this if you want to re-run.".format(
                    modelcomparisons[gene][cutoff]))
            continue
        print("\nAnalyzing {0} using homologs with identity >= {1}...".format(gene, cutoff))
        prefsfiles = prefs[gene].values()
        cmd = 'phydms_comprehensive {0} {1} {2} '.format(outprefix, alignment, ' '.join(prefsfiles)) +\
                 '--yngkp M3 --no-avgprefs --randprefs --no-omegabysite --no-diffprefsbysite --no-stringencybysite'
        ncpus = len(prefsfiles) * 2 + 1
        print("Running the following command via slurm:\n" + cmd)
        # rather than just running *cmd* directly, we submit it via sbatch
        sbatchfile = outprefix + '.sbatch'
        with open(sbatchfile, 'w') as f:
            f.write('#!/bin/bash\n' + cmd)
        subprocess.check_call(['sbatch', '--cpus-per-task', str(ncpus), '--time', '48:00:00', sbatchfile])
        os.remove(sbatchfile)
```

### Comparing the model fits
Now let's look at how well the various models fit the data for each of the four sets of orthologs. We can compare model fit using [Aikake Information Criterion (AIC)](https://en.wikipedia.org/wiki/Akaike_information_criterion), which compares the likelihoods of the data giving the model after penalizing for the number of free parameters. 

For each set of model comparisons shown by the cell below, the models are sorted from best (lowest AIC) to worst (highest AIC). Note that *ExpCM* denotes *experimentally informed codon model*, and *YNGKP* indicates the Goldman-Yang style standard substitution model.

```python
for gene in genes:
    for cutoff in identity_cutoffs:
        fname = modelcomparisons[gene][cutoff]
        print("\nModel comparisons for {0} using orthologs with >= {1} protein identity.".format(gene, cutoff))
        print("This is file {0}".format(fname))
        with open(fname) as f:
            display(HTML('<font size=1>' + docutils.core.publish_string(f.read(), writer_name='html') + '</font>'))
```

### Interpreting the model comparisons
The tables above list a lot of entries, but the results are quite straightforward. For all four sets of homologs, all of the models informed by the deep mutational scanning data are much better than standard Goldman-Yang models or models in which the experimental data has been randomized among sites. This means that the experimental measurements are definitely capturing some of the real evolutionary constraints.

Second, the results of the model comparisons are similar regardless of whether we use the orthologs with at least 85% or at least 75% identity. This means that the shifts in mutational effects that occurred during the evolutionary divergence of the human SUMO1 and UBE2I from the other orthologs are sufficiently small that the experiemnts on the human variants provide broadly useful information. Given this, it seems preferable to look mostly at the sets of orthologs with at least 75% identity, since these sets are larger and provide more evolutionary information due to the higher divergence. So for the rest of this analysis, we will look at the results from these sets.

Looking at these sets, we get a clear answer of how we should interpret the fitness effects of mutations that enhance activity in the experiment: increases in activity should be thought of as evolutionary deleterious. For both UBE2I and SUMO1, penalizing increased activity yields a better model. Treating variants with increased activity as equivalent to wildtype is second best. And treating variants with increased activity as having higher fitness is the worst. 

Finally, what are the relative benefits of including or excluding the conservation data when generating the joint scores? For SUMO1, including the conservation data doesn't help at all in terms of making the measurements better reflect natural evolutionary constraints (actually, it makes thing *very slightly* worse). For UBE2I, including the conservation data is helpful, although the model without these data is also quite good. But given the concerns about the circularity of using conservation data (which is generated from analyzing the orthologs themselves), from here on we will focus on the models without conservation data.

### Interpreting differences in the stringency of selection in nature and in the experiments
Another very informative parameter from the model fitting is the *stringency parameter*, $\beta$. Essentially, this parameter tells us the extent that natural selection prefers the same mutations as the experiments. If $\beta > 1$, then natural selection prefers the same mutations as the experiments, but even more so. If $\beta < 1$, then natural selection prefers the mutations favored in the experiments relatively less. More precisely, the preferences are re-scaled to be proportional to $\left(\pi_{r,a}\right)^{\beta}$. Note that under certain assumptions, $\beta$ actually has a population-genetic interpretation as effective population size [Bloom 2016](http://biorxiv.org/content/early/2016/01/22/037689). In other deep mutational scanning data sets, usually $\beta > 1$ which means that the deep mutational scanning is indeed identifying favorable and unfavorable mutations, but that natural selection does so with much greater stringency; see [Bloom 2016](http://biorxiv.org/content/early/2016/01/22/037689) .

The tables above show that the same is true for the experiments on SUMO1 and UBE2I, which fit $\beta = 3.20$ and $\beta = 1.34$ on the models that we have chosen for consideration (orthologs with at least 75% identity, no-conservation scores, penalize activity-enhancing mutations). Reassuringly, when data are randomized among sites, $\beta$ becomes close to zero (essentially flattening the data). So these results show that natural selection likes the same amino acids as the experiments, but even more so.

### Visualizing the deep mutational scanning data
We can use logo plots to visualize the site-specific amino-acid preferences from the deep mutational scanning data re-scaled by the optimal stringency parameters described above. Here are those plots, rendered using [dms_logoplot](http://jbloomlab.github.io/dms_tools/dms_logoplot.html).

Recall that we are using the no-conservation scores with the fitting done by penalizing mutations with enhanced activity and using the orthologs with at least 75% identity.

```python
print("Using " + subprocess.check_output(['dms_logoplot', '-v'], stderr=subprocess.STDOUT))

logoplotsdir = './logoplots/'
if not os.path.isdir(logoplotsdir):
    os.mkdir(logoplotsdir)

# get the models we decided are best for our purposes
bestcutoff = 0.75 # use this identity cutoff
bestscore = 'no-cons_penalize-scores-gt1' # use this deep mutational scanning data
bestprefix = {} # prefix for files for 
for gene in genes:
    bestprefix[gene] = prefixes[gene][cutoff] + '_ExpCM_' + gene + '_' + bestscore + '_prefs'
    with open(bestprefix[gene] + '_modelparams.txt') as f:
        beta = float([line.split('=')[1] for line in f.readlines() if 'stringencyparameter' in line][0])
    print("\nFor gene {0}, the stringency parameter is {1}".format(gene, beta))
    iprefs = prefs[gene][bestscore]
    assert os.path.isfile(iprefs), iprefs
    print("Here is a logo plot of the deep mutational scanning data re-scaled to account for stringency differences:")
    prefsplot = logoplotsdir + gene + '_prefs.pdf'
    subprocess.check_call([
            'dms_logoplot',
            iprefs,
            prefsplot,
            '--stringencyparameter', str(beta),
            '--nperline', '53',
            ])
    ShowPDF(prefsplot)
```

### Looking for sites of selection
We can use [phydms](http://jbloomlab.github.io/phydms/phydms_prog.html) to look for sites of diversifying or differential selection. Sites under diversifying selection are changing faster than expected in nature given the deep mutational scanning data. Sites under differential selection prefer different amino acids in nature than expected given the deep mutational scanning data.

We do this analysis for SUMO1 and UBE2I using the alignments of orthologs with at least 75% identity, and the no-conservation scores that penalize activity-enhancing mutations. The code block below runs the analysis, and then plots the results in a visual form.

```python
# do analysis to get differential and diversifying selection
for gene in genes:
    cmd = [
            'phydms',
            alignments[gene][bestcutoff],
            bestprefix[gene] + '_tree.newick',
            'ExpCM_' + prefs[gene][bestscore],
            bestprefix[gene],
            '--no_optimize',
            '--diffprefsbysite',
            '--omegabysite',
            '--ncpus', '-1'
            ]
    diffsel = bestprefix[gene] + '_diffprefsbysite.txt'
    divsel = bestprefix[gene] + '_omegabysite.txt'
    if all(map(os.path.isfile, [diffsel, divsel])):
        print("\nPer site-selection analyses have already been run for {0}".format(gene))
    else:
        print("\nDetecting per-site selection for {0} with command:\n{1}".format(gene, ' '.join(cmd)))
        subprocess.check_call(cmd)
    selectionplot = logoplotsdir + gene + '_selection.pdf'
    print("Now plotting this selection in {0}".format(selectionplot))
    subprocess.check_call(['phydms_plotselection', diffsel, selectionplot, '--omegabysite', divsel, '--nperline', '53'])
    ShowPDF(selectionplot)

                          
```

### Interpreting the diversifying and differential selection
The plots above show that there are no sites under particularly strong diversifying selection, which is probably expected as SUMO1 and UBE2I are probably not under repeated pressure for change due a genetic or host-virus conflict.

However, there are some sites (e.g. positions 1-3 in SUMO1, position 155 in UBE2I, and some others) where natural selection clearly favors different mutations than the deep mutational scanning. There is probably some biological meaning to this, but I don't know enough about the genes to make an interpretation... I can note that most of these are sites that are *not* under much constraint in the experiments (preferences logo plots a few cells above), but do appear to be under greater constraint for specific amino acids in nature.

## Conclusions
The deep mutational scanning experiments clearly capture information that reflects the natural evolutionary constraints on orthologs of SUMO1 and UBE2I.

The best way to treat activity-enhancing mutations is as deleterious -- namely, make their preference inversely proportional to the increase in activity. This suggests that over-activating and well as inactivating mutations are harmful during natural evolution.

Although the experiments describe the general constraints on both genes well, there are some sites under different selection pressures in nature versus the lab. I don't know enough about the genes to determine the biological interpretation of what is happening at these sites.
